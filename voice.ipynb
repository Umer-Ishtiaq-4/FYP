```python

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device {device}")
bundle = torchaudio.pipelines.WAV2VEC2_ASR_BASE_960H
print(f"Sample Rate: {bundle.sample_rate}")
print(f"Labels: {bundle.get_labels()}")
model = bundle.get_model().to(device)
print(model.__class__)
```

```python
#waveform, sample_rate = torchaudio.load(SPEECH_FILE)
waveform, sample_rate = torchaudio.load('test.wav')
waveform = waveform.to(device)
if sample_rate != bundle.sample_rate:
    waveform = torchaudio.functional.resample(waveform, sample_rate, bundle.sample_rate)
```

```python
with torch.inference_mode():
    emission, _ = model(waveform)
```

```python
class GreedyCTCDecoder(torch.nn.Module):
    def __init__(self, labels, blank=0):
        super().__init__()
        self.labels = labels
        self.blank = blank

    def forward(self, emission: torch.Tensor) -> str:
        indices = torch.argmax(emission, dim=-1) # [num_seq,]
        indices = torch.unique_consecutive(indices, dim=-1)
        indices = [i for i in indices if i != self.blank]
        return "".join([self.labels[i] for i in indices])
```

```python
decoder = GreedyCTCDecoder(labels=bundle.get_labels())
transcript = decoder(emission[0])
```

```python
import spacy
from spacy import displacy


nlp = spacy.load("en_core_web_sm")


sentence = "a bedroom next to the kitchen and a lounge next to the dining room"


doc = nlp(sentence)


displacy.render(doc, style='dep', jupyter=True, options={})


for token in doc:
    for child in token.children:
        print(token.text, "->", child.dep_, "->", child.text)
```

```python
import spacy

# Load the language model
nlp = spacy.load("en_core_web_sm")

# Input sentence
sentence = "a bedroom next to the kitchen"

# Process the sentence using the language model
doc = nlp(sentence)

# Extract relations between words and parts of speech
for token in doc:
    for child in token.children:
        print(token.text, "->", token.pos_, "->", child.dep_, "->", child.text, "->", child.pos_)
```
